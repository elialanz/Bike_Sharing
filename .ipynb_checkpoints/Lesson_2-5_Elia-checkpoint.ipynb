```python
# Load the CSV file into a dataframe and use the first column as the index

df = pd.read_csv('chicago_data.csv', index_col = 0)
```

## Data preprocessing

```python
# Create a value column and group by start and end station
# Add a helper column and count trips for each start–end station pair using a groupby operation

df['value'] = 1
df_group = df.groupby(['from_station_name', 'to_station_name'])['value'].count().reset_index()
```

### Some exploratory analysis into the trips

```python
# Plot a histogram of trip counts using 100 bins

df_group['value'].hist(bins = 100)
plt.show()
```

```python
# Create and display a boxplot showing the distribution of trip count values

b_plot = df_group.boxplot(column = 'value') 
b_plot.plot()
```

```python
# Count how many times each trip count value appears, sorted by index

df_group['value'].sort_index().value_counts()
```

```python
# Export the grouped dataframe to a CSV file

df_group.to_csv('df_groupby_final.csv')
```

```python
# Display the first five rows of the grouped dataframe

df_group.head()
```

```python
# Rename station and value columns to clearer names directly in the dataframe

df_group.rename(columns = {'from_station_name':'start_station_name','to_station_name' : 'end_station_name',
                          'value': 'trips'}, inplace = True)
```

## Merge data to create latitude and longitude - only for Chicago

#### Import location data - only for Chicago

```python
# Load the station locations CSV file into a dataframe using the first column as the index
# Source: https://data.cityofchicago.org/Transportation/Divvy-Bicycle-Stations/bbyy-e7gq

df_stations = pd.read_csv("Divvy_Locations.csv", index_col = 0)
```

```python
# Display the first five rows of the stations dataframe

df_stations.head()
```

```python
# Create a new start_station_name column by copying values from Station Name

df_stations['start_station_name'] = df_stations['Station Name']
```

```python
# Display the first five rows of the stations dataframe

df_stations.head()
```

```python
# Rename the Station Name column to end_station_name directly in the dataframe

df_stations.rename(columns = {'Station Name': 'end_station_name'}, inplace = True)
```

```python
# Reset the dataframe index and turn it into a regular column

df_stations.reset_index(inplace = True)
```

```python
# df_stations = df_stations[['end_station_name', 'start_station_name', 'Latitude', 'Longitude']]

df_stations = df_stations[['end_station_name', 'start_station_name', 'Latitude', 'Longitude']]
```

```python
# Display the cleaned stations dataframe

df_stations
```

##### _By start station_

```python
# Merge trip counts with station location data using start station name and keep merge status flags

df_m = df_group.merge(df_stations, how = 'outer', on = "start_station_name", indicator = 'merge_flag')
```

```python
# Count how many rows fall into each merge category, including unmatched values

df_m['merge_flag'].value_counts(dropna = False)
```

```python
# Filter the merged dataframe to keep only rows that matched in both tables

df_m = df_m[df_m['merge_flag'] =='both']
```

```python
# Display the number of rows and columns in the merged dataframe

df_m.shape
```

```python
# Display the first five rows of the merged dataframe

df_m.head()
```

```python
# Remove the duplicate end station name column from the merged dataframe

df_m.drop(columns = {'end_station_name_y'}, inplace = True)
```

```python
# Rename the remaining end station name column to end_station_name

df_m.rename(columns = {'end_station_name_x' : 'end_station_name'}, inplace = True)
```

##### _By end station_

```python
# Merge the dataframe again with station data using end station name and track merge results

df_final = df_m.merge(df_stations, how = 'outer', on = "end_station_name", indicator = 'merge_flag_2')
```

```python
# Count how many rows fall into each second merge category, including unmatched values

df_final['merge_flag_2'].value_counts(dropna = False)
```

```python
# Display the first five rows of the final merged dataframe

df_final.head()
```

```python
# Filter the final dataframe to keep only rows that matched in both tables

df_final = df_final[df_final['merge_flag_2'] =='both']
```

```python
# Remove duplicate station name and merge flag columns from the final dataframe

df_final.drop(columns = {'start_station_name_y', 'merge_flag', 'merge_flag_2'}, inplace = True)
```

```python
# Rename the remaining start station name column to start_station_name

df_final.rename(columns = {'start_station_name_x' : 'start_station_name'}, inplace = True)
```

```python
# Display the first five rows of the final cleaned dataframe

df_final.head()
```

```python
# Force Python to run garbage collection and free up unused memory

import gc # this is garbage collector (speeds up performance)
gc.collect()
```

```python
# Display the first five rows of the final cleaned dataframe

df_final.head()
```

```python
# Rename latitude and longitude columns for start and end station locations

df_final.rename(columns = {'Latitude_x' : 'start_lat', 'Longitude_x' : 'start_lon', 
       'Latitude_y' : 'end_lat', 'Longitude_y' : 'end_lon',}, inplace = True)
```

## Plot The Map

```python
# Export the final dataframe with station locations to a CSV file

df_final.to_csv('df_final_locations_for_map.csv')
```

```python
# Create a Kepler.gl interactive map using the final dataframe and set the map height to 700

m = KeplerGl(height = 1000, data={"data_1": df_final})
m
```

```python
# Store the current Kepler.gl map configuration in a variable

config = m.config
```

```python
# Display the current Kepler.gl map configuration stored in the config variable

config
```

```python
# Save the Kepler.gl map configuration to a JSON file named config.json

import json
with open("config.json", "w") as outfile:
    json.dump(config, outfile)
```

```python
# Save the interactive Kepler.gl map to an HTML file with the specified configuration

m.save_to_html(file_name = 'Divvy Bike Trips Aggregated.html', read_only = False, config = config)
```
